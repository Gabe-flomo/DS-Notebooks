```python
import pandas as pd
```

```python
def load_data(path = r"E:\Documents\Data\Datasets"):
    # point to the data set directory and choose the file to load
    import os
    p = path
    os.chdir(p) 

    #choose a data set
    while True:
        files = []
        with os.scandir(p) as dir:
            for count,entry in enumerate(dir):
                print(f"{count}) {entry.name}")
                files.append(entry.name)




        data = int(input('Enter the file index: '))
        data = os.path.join(p,files[data])
        if not data.endswith('.csv'):
            p = data
            print("\nThis is a directory\n")
        else:
            print(data)
            break
            
    return data
```

```python
data1 = load_data(r"E:\Documents\Data\Datasets")
data2 = load_data(r"E:\Documents\Data\Datasets")
```

# Grouping
If we want to see specific results based on some column,then were going to have to group on that specific column. For example, if we want to see what gender prefers specific social media sites, we'll have to group by gender.

### what are the most popular social media sites for each country?
In this case we have to group on the country column

```python
# first lets see the value counts for each country
df['Country'].value_counts(normalize = True).head(10) * 100
```

This shows that __23%__ of the data comes from the US, __10%__ of the data comes from India, __6%__ comes from Germany and so on.

```python
# group by column(s) by passing a list of the columns you want to create a group of, then it returns a groupby object 
country_group = df.groupby(['Country'])
```

```python
# in this case were grouping by country and by student
country_student_group = df.groupby(['Country','Student'])
```

```python
# now we want to look at the median salary per country, to do this all we have to do is use our country_group and apply
# the median function to the ConvertedComp column, which is just the salary in USD.
country_group['ConvertedComp'].median()
```

```python
# we can use the query method to check for a condition, in this case where the mean salary is less than $100,000
country_group['ConvertedComp'].agg(['mean','median']).query("mean > 150000")
```

```python
# next were going to use a filter to get the entries in the data frame and check whether or not the respondent from the
# specified country uses python. This will return a series of true and false values on whether or not the respondent in that
# country uses python or not.
filt = df['Country'] == 'India'
df[filt]['LanguageWorkedWith'].str.contains('Python')
```

```python
# if we want to see the number of people who know python instead of true/false values, we can use the sum method which
# shows that 3,105 people in india work with python
df[filt]['LanguageWorkedWith'].str.contains('Python').sum()
```

# The Apply function
since our country_group is a series groupby object, we wont be able to use the .str method to see which respondents know python. Instead were going to have to use the .apply method. When using the apply method, were going to specify a function that we want to apply to every value in the series. In this case x is the series (Language worked with) and we are going to see if x contains python and get the sum total.

[Docs on apply](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.apply.html)

[examples](https://www.journaldev.com/33478/pandas-dataframe-apply-examples)

[more examples](https://www.geeksforgeeks.org/python-pandas-apply/)

[lambda functions](https://www.w3schools.com/python/python_lambda.asp)

```python
country_group['LanguageWorkedWith'].apply(lambda x: x.str.contains("Python").sum())
```

```python
# and if i add .loc at the end, I can look for a specific country
country_group['LanguageWorkedWith'].apply(lambda x: x.str.contains("Python").sum()).loc["United States"]
```

```python
# now if we want to see those values in percents, so what is the percent of developers in each country that use python
country_group['LanguageWorkedWith'].apply(lambda x: x.str.contains("Python").mean()*100)
# or 
country_group['LanguageWorkedWith'].apply(lambda x: x.str.contains("Python").sum()/x.count()*100)
```

```python
# Another method is to count the number of respondents from each country and make that a series, then count the number 
# of respondents that use python and turn that into a series. Then we can concat those series together to make a new
# dataframe. 
number_of_respondents = df['Country'].value_counts()
uses_python = country_group['LanguageWorkedWith'].apply(lambda x: x.str.contains("Python").sum())
# we set the axis to columns because by default it will want to concatonate on the rows, which will just make our dataframe
# into one long series. 
python_df = pd.concat([number_of_respondents,uses_python],axis = 'columns',sort = False)
python_df.rename(columns = {'Country':'Respondants','LanguageWorkedWith':'Uses_Python'},inplace = True)
python_df
```

```python
# now were going to make a new column for the percentage of people that know python. Percentage == (part/whole) * 100
python_df['pct_python_users'] = (python_df['Uses_Python']/python_df['Respondants']) * 100
python_df
```

```python
# sorting the dataframe by highest percents
python_df.sort_values(by = 'pct_python_users',ascending = False)
```

```python
# same thing but only for countries where the respondant count is greater than 100 and the percent is higher than 0
python_df.sort_values(by = 'pct_python_users',ascending = False).query('Respondants > 100 and pct_python_users > 0').head(15)
```

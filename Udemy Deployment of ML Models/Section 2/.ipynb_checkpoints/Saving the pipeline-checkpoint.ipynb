```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error
from math import sqrt
import joblib
import warnings
warnings.simplefilter(action='ignore')
pd.pandas.set_option('display.max_columns',None)
import os
def load_data(path):
    # point to the data set directory and choose the file to load
    p = path
    os.chdir(p) 

    #choose a data set
    while True:
        files = []
        with os.scandir(p) as dir:
            for count,entry in enumerate(dir):
                print(f"{count}) {entry.name}")
                files.append(entry.name)




        data = int(input('Enter the file index: '))
        data = os.path.join(p,files[data])
        if not data.endswith('.csv'):
            p = data
            print("\nThis is a directory\n")
        else:
            print(data)
            break
            
    return data

path = r'E:\Documents\Data\Datasets'
```

# Handle Missing values
### Categorical variables
For categorical variables, we will replace missing values with the string "missing".

```python
NA = [var for var in features if xtrain[var].isnull().sum()>0 and xtrain[var].dtypes == 'O']
NA
```

```python
xtrain[NA] = xtrain[NA].fillna('Missing')
xtest[NA] = xtest[NA].fillna('Missing')
xtrain[NA].isnull().sum()
```

### Numerical variables

To engineer missing values in numerical variables, we will:

- add a binary missing value indicator variable
- and then replace the missing values in the original variable with the mode


```python
NA = [
    var for var in features
    if xtrain[var].isnull().sum() > 0 and xtrain[var].dtypes != 'O'
]

NA
```

```python
mode = xtrain['LotFrontage'].mode()[0]
mode
```

```python
xtrain['LotFrontage'] = xtrain['LotFrontage'].fillna(mode)
xtest['LotFrontage'] = xtest['LotFrontage'].fillna(mode)
```

## Temporal variables

One of our temporal variables was selected to be used in the final model: 'YearRemodAdd'

So we need to deploy the bit of code that creates it.

```python
def elapsed(df, var):
    ''' capture the difference between the var and year sold '''
    df[var] = df['YrSold'] - df[var]
    return df
```

```python
xtrain = elapsed(xtrain, 'YearRemodAdd')
xtest = elapsed(xtest, 'YearRemodAdd')
```

# Numerical variable transformation

```python
# apply a log transform to the variables that were selected and the target
# natural log, inverse = e (np.exp)
for var in ['LotFrontage', '1stFlrSF', 'GrLivArea', 'SalePrice']:
    xtrain[var] = np.log(xtrain[var])
    xtest[var] = np.log(xtest[var])
    
```

# Categorical Variables
#### Group rare labels

```python
cat = [var for var in features if xtrain[var].dtype == 'O']
cat
```

```python
def frequent(df, var, rarity = .01):
    # function finds the labels that are shared by more than
    # a certain % of the houses in the dataset
    df = df.copy()
    tmp = df.groupby(var)['SalePrice'].count() / len(df) 
    return tmp[tmp>rarity].index

frequent(xtrain,'Neighborhood')
```

```python
for var in cat:
    
    #find frequent labels
    frequent_list = frequent(df,var)
    print(var)
    print(frequent_list)
    print()
    
    # replace rare categories with the string rare
    xtrain[var] = np.where(xtrain[var].isin(frequent_list),
                          xtrain[var], # if not in the list keep the category string
                          'Rare')      # if in the list replace with rare
    
    xtest[var] = np.where(xtest[var].isin(frequent_list),
                         xtest[var],
                         'Rare')
```

# Encoding Categorical Variables

```python
# this function will assign discrete values to the strings of the variables,
# so that the smaller value corresponds to the category that shows the smaller
# mean house sale price

def encode(train, test, var, target):
    # order the categories in a variable from that with the lowest
    # house sale price, to that with the highest
    ordered_labels = train.groupby([var])[target].mean().sort_values().index
    
    # create a dictionary of ordered categories to integer values
    ordinal_labels = {k:i for i,k in enumerate(ordered_labels,0)}
    
    # use the dictionary to replace the categorical strings by integers
    train[var] = train[var].map(ordinal_labels)
    test[var] = test[var].map(ordinal_labels)
    print(var)
    print(ordinal_labels)
    print()
```

```python
for var in cat:
    encode(xtrain,xtest,var,"SalePrice")
```

```python
# for handling data
import numpy as np
import pandas as pd

# for visualization
import matplotlib.pyplot as plt
%matplotlib inline

# misc
import os
```

```python
def load_data(path):
    # point to the data set directory and choose the file to load
    p = path
    os.chdir(p) 

    #choose a data set
    while True:
        files = []
        with os.scandir(p) as dir:
            for count,entry in enumerate(dir):
                print(f"{count}) {entry.name}")
                files.append(entry.name)




        data = int(input('Enter the file index: '))
        data = os.path.join(p,files[data])
        if not data.endswith('.csv'):
            p = data
            print("\nThis is a directory\n")
        else:
            print(data)
            break
            
    return data
```

# missing values

```python
# this list comprehension makes a list of all the variables that contain missing values
na_var = [var for var in df.columns if df[var].isnull().sum()>1]

# print the variable name and the percentage of missing values
for var in na_var:
    # finds the column in the dataset with df[var]
    # counts how many observations are null
    # converts the amount of null values to a percent by calculating the mean
    # then rounds to 3 decimal places
    print(f"{var} has {np.round(df[var].isnull().mean()*100, 3)}% missing values")
```

 # Relationship between missing values and sales price 
 ### if a value is missing, does it effect the sales price?
 [docs on np.where()](https://numpy.org/devdocs/reference/generated/numpy.where.html)

```python
def analyze_na(df,var):
    # create a copy of the dataframe
    df = df.copy()
    
    # makes the column show a 1 if the value is missing and a 0 otherwise
    # so now we have a binary variable that determines if a value was missing or not
    df[var] = np.where(df[var].isnull(),1,0)
    
    # this groups the functions by either the 0s or the 1s then calculates
    # the median house price and displays it as a plot
    df.groupby(var)['SalePrice'].median().plot.bar()
    # show the column name as the title
    plt.title(var)
    plt.show()
    
# loop over all the variables show a plot for the missing values
for var in na_var:
    analyze_na(df,var)
```

Here we can see that the median house price is less when the values are missing. This means that the fact that the info is missing, it will probably effect the final price. When a value is present the houses are more expensive.

# Recap so far
what we just did above was find all of the features that had missing values. Then we asked the question: 

_"does missing values effect the price of the median price of the house?"_

To answer this question we want to group by observations with missing values, and observations without missing values and look at the median price for each, per feature.

#### Temporal variables

We have 4 year variables in the dataset:

- YearBuilt: year in which the house was built
- YearRemodAdd: year in which the house was remodeled
- GarageYrBlt: year in which a garage was built
- YrSold: year in which the house was sold

We generally don't use date variables in their raw format. Instead, we extract information from them. For example, we can capture the difference in years between the year the house was built and the year the house was sold.

```python
# get the columns that contain year info
years = [var for var in numerical if 'Year' in var or 'Yr' in var]

years
```

```python
# print unique values for year data
for var in years:
    print(f"{var} {df[var].unique()}")
    print()
```

```python
# plot the median house price over the years (year sold)
df.groupby('YrSold')['SalePrice'].median().plot()
plt.ylabel('Median house Price')
plt.title('Change in house price over the years');
```

As time goes by and houses become older, their tends to decrease. This information can be well indicated in an interval variable.

```python
def analyze_year(df,var):
    df = df.copy()
    
    # capture the difference between a year feature like when it was
    # remodeled and the year the house was sold.
    # This basically shows us, as things were done to the house 
    # (eg. remodeling, when it was built, added a garage, etc.)
    # how much time passed between then and when it was sold and how that
    # time effects the median house price.
    df[var] = df['YrSold'] - df[var]
    
    plt.scatter(df[var], df['SalePrice'])
    plt.ylabel('Sales Price')
    plt.xlabel(var)
    # this will show individual graphs
    plt.show()
    
for year in years:
    if year != 'YrSold':
        analyze_year(df,year)
```

Here we can see a trend that as more time passes from each feature and the year that it was sold, the median price tends to decrease. This also suggest that these variables could be very predictive.

# Discrete Variables

```python
# here discrete values are assumed to be features where the number of unique variables is less than 20 and the variable isnt a year variable
discrete = [var for var in numerical if len(df[var].unique()) < 20 and var not in years]
```

```python
print(f"Number of discrete variables: {len(discrete)}")
```

```python
df[discrete].head()
```

```python
def analyze_dis(df, var):
    df = df.copy()
    # this will group on each feature and plot the median
    # house price based on each discrete variable
    df.groupby(var)['SalePrice'].median().plot.bar()
    plt.title(var)
    plt.ylabel('SalePrice')
    plt.show()
    
for dis in discrete:
    analyze_dis(df, dis)
```

There tend to be a relationship between the variables values and the SalePrice, but this relationship is not always monotonic.

For example, for OverallQual, there is a monotonic relationship: the higher the quality, the higher the SalePrice.

However, for OverallCond, the relationship is not monotonic. Clearly, some Condition grades, like 5, correlate with higher sale prices, but higher values do not necessarily do so. We need to be careful on how we engineer these variables to extract maximum value for a linear model.

# Analyzing the Distributions of continuous variables

we want the relationship of the continuous variables to be gaussian instead of skewed for linear models.

```python
def analyze_cont(df, var):
    df = df.copy()
    df[var].hist(bins = 30)
    plt.ylabel('Number of Houses')
    plt.xlabel(var)
    plt.title(var)
    plt.show()
    
for cont in continuous:
    analyze_cont(df, cont)
```

Here we can see that most of the data has a skewed distribution.

```python
def cont_transform(df, var):
    df = df.copy()
    
    # log doesnt take or negative values so we can skip those
    if 0 in df[var].unique():
        print(var)
    else:
        df[var] = np.log(df[var])
        df[var].hist(bins = 20)
        plt.ylabel('Number of houses')
        plt.xlabel(var)
        plt.title(var)
        plt.show()
        
for cont in continuous:
    cont_transform(df,cont)
```

Now the variables follow more of a normal distribution.

# Explore the relationship betweenthe house price and the transformed variables

```python
def transform_relation(df, var):
    df = df.copy()
    
    if 0 in df[var].unique():
        pass
    else:
        # our x-axis
        df[var] = np.log(df[var])
        # our y-axis
        df['SalePrice'] = np.log(df['SalePrice'])
        plt.scatter(df[var], df['SalePrice'])
        plt.ylabel('Sale Price')
        plt.xlabel(var)
        plt.title(f"Correlation of sales price vs {var}")
        plt.show()

for cont in continuous:
    if cont != 'SalePrice':
        transform_relation(df, cont)
```

Here we can see that the first few plots dont show much correlation, but the last 2 seem to have a positive correlation. As the x value increases the sales price also increases.

# outliers
we can visualize the outliers with a boxplot

```python
def outliers(df, var):
    df = df.copy()
    
    if 0 in df[var].unique():
        pass
    else:
        df[var] = np.log(df[var])
        df.boxplot(column = var)
        plt.title(var)
        plt.ylabel(var)
        plt.show()

# outliers are the values that are shown beyond the lines
for cont in continuous:
    outliers(df, cont)
```

# Rare labels
labels that appear in a small portion of the dataset

```python
def rare(df, var, rarity = .01):
    df = df.copy()
    # this is the percent that each category shows up 
    tmp = df.groupby(var)['SalePrice'].count() / len(df)
    return tmp[tmp<rarity]

for cat in categorical:
    print(rare(df, cat))
    print()
```
